### Analisis Supermercados de EEUU

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./supermercados.csv")
df
```

```python
# como tenemos valores NaN "nulos" hay que cambiarlos a 0
df.fillna(0, inplace=True)

# verificamos si ya cambiaron los datos nulos
df
```

```python
# Datos Types for our columns

print(df.dtypes)
```

```python
# Cambiar type a integer de la columna stores
df["Stores"] = df['Stores'].astype("int64")
```

### 2. ¿Cuánto dinero en ventas generó la mayoría de las compañías?

```python
plt.figure(figsize=[16,8])
df_ordenada = df.sort_values('Sales', ascending=0)
x = df_ordenada['Company']
y = df_ordenada['Sales']

plt.bar(x,y)
plt.title('Number of Sales')
plt.xlabel('Company')
plt.ylabel('Sales')
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=[16,8])
df[1:]['Sales'].plot.hist(bins=4)
df['Sales'].median()
```

R.2 = En la primera gráfica se observa la disparidad en ventas que tiene Waltmart con respecto a las otras companias (era la finalidad de realizarla) mientras que en la segunda se observa que realmente la mayoría de las companias generan en ventas valores menores a 40 m usd. Por esta razon creo conveniente el calculo de la mediana para la respuesta de la pregunta, la cual seria que la mayor cantidad de empresas generan ventas de aproximadamente 34980 mill usd

### 3. ¿Cuantas tiendas tiene la mayoria de las compañias?

```python
df_ordenados = df.sort_values(by='Stores', ascending=False)
df_ordenados
```

```python
sns.histplot(data = df_ordenados, x='Stores', color = 'orange', bins=12,)
sns.set(rc = {'figure.figsize':(16,8)})
df_ordenados[0:23]['Stores'].median()

```

R.3 = Al observar la gráfica se puede ver claramente que el numero de tiendas de la mayoría de tiendas es menor a 2000, cantidad que se puede corroborar con el calculo de la mediana , el cual como se observa, es igual a 1828 De este calculo vale la pena decir que 2 de las 25 empresas no poseen tiendas físicas por lo que lo ideal es dejarlas fuera de la ecuación.

## 4. ¿La cantidad de tiendas está relacionada con la cantidad de ventas? Es decir, ¿si una empresa tiene más tiendas tendrá más dinero de ventas?

R.4 = En la gráfica anterior podemos visualizar que la compania dominante (en términos de numero de tiendas) a pesar de poseer mas de 14 mil tiendas sus ventas no superan los 50mil millones de dolares, mientras que por otra parte Walmart que es indiscutiblemente la empresa dominante en cuanto a ventas, apenas posee menos de 6000 mil tiendas físicas.

## II. Preguntas opcionales

### 7. ¿Qué categoría de compañía generó más ventas?

```python
company = df.groupby(by=['Category'])['Sales'].sum()

sns.scatterplot(data=company)
```

R.7 = Se puede observar claramente en la gráfica que la categoría de ventas Supercenters tiene ventas por encima de los 700.000 millones de dolares convirtiéndola así en la categoría que mas genera dinero de las que se están estudiando.

### 8. ¿Cuál es la compañía que en relación con su cantidad de tiendas físicas genera más ventas?

```python
sales_stores = df.sort_values(by='Stores', ascending=True)
sns.scatterplot(data=sales_stores, x = 'Sales/Avg. Store', y = 'Company', hue='Stores',s=200)
```

R.8 = Se observa que la empresa que mas produce ganancias por menor numero de tiendas físicas es Costco.
